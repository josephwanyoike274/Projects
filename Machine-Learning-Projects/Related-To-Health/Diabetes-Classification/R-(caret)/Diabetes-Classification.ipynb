```r
# Set the seed for reproducibility
set.seed(123)
```

```r
# Split the data into training and testing sets
trainIndex <- createDataPartition(data$Diabetic, p = 0.7, list = FALSE)
trainData <- data[trainIndex, ]
testIndex <- setdiff(1: nrow(data), trainIndex)
testData <- data[testIndex,]
```

```r
# Define the control parameters for the model
control <- trainControl(method = "cv", number = 10)
```

```r

# Convert target variable to a factor with two levels
trainData$Diabetic <- factor(trainData$Diabetic)
```

```r
# Train the logistic regression model
model <- train(Diabetic ~ ., data = trainData, method = "glm", family = "binomial", trControl = ctrl)
```

```r
print(model)
```

```r
# Make predictions on the test set
predictions <- predict(model, newdata = testData)
#predictions <- factor(predictions, levels = levels(testData$Diabetic))
```

```r
# Evaluate the model
confusionMatrix(predictions, testData$Diabetic)
```

```r
identical(levels(predictions), levels(testData$Diabetic))
```

```r
levels_predictions <- levels(predictions)
levels_testData <- levels(testData$Diabetic)
```

```r
missing_levels <- setdiff(levels_testData, levels_predictions)
```

```r
predictions <- factor(predictions, levels = c(levels_predictions, missing_levels))
```
